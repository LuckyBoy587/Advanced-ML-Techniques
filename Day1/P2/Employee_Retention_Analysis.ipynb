# Employee Retention Analysis based on Salary

This notebook analyzes the relationship between employee salary levels and retention outcomes.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'https://github.com/LuckyBoy587/Advanced-ML-Techniques/raw/refs/heads/master/Day1/P2/ML470_S1_HR_Data_Practice.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows
df.head()
```

```python
# Create a bar plot to visualize the impact of salary on employee retention
df.groupby("salary")["left"].value_counts().unstack().plot(kind='bar')
plt.title('Impact of Salary on Employee Retention')
plt.xlabel('salary')
plt.ylabel('Count')
plt.show()
```
