```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("ML470_S6_Mobile_Data_Practice.xlsx")
df.head()
```

```python
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
numerical_features.remove('price_range')  

fig, axes = plt.subplots(figsize=(16, 6))
df[numerical_features].boxplot(ax=axes)
plt.xticks(rotation=45,)
plt.ylabel('Value')
plt.title('Boxplots of All Numerical Input Features - Outlier Assessment')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, fmt=".3f", cmap='YlGnBu', cbar_kws={'shrink': 0.8})
plt.tight_layout()
plt.show()
```
