```python
import pandas as pd

df = pd.read_excel(r"https://github.com/LuckyBoy587/Advanced-ML-Techniques/raw/refs/heads/master/Day3/P2/ML470_S3_HR_Data_Practice.xlsx")
df.head()
```

```python
from sklearn.calibration import LabelEncoder


le = LabelEncoder()
df["salary.enc"] = le.fit_transform(df["salary"])
df["Department.enc"] = le.fit_transform(df["Department"])
df = df.drop(columns=["salary", "Department"])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=45)
sns.boxplot(df)
```
