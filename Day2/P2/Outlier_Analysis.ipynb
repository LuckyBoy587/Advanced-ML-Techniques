# Employee Retention Analysis: Outlier Detection

## Objectives
1. Detect outliers in the dataset.
2. Visualize the outliers using boxplots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the dataset from raw GitHub URL
url = "https://raw.githubusercontent.com/LuckyBoy587/Advanced-ML-Techniques/master/Day2/P2/ML470_S2_Practice_HR_Data.xlsx"
try:
    df = pd.read_excel(url)
    print("Dataset loaded successfully.")
    display(df.head())
except Exception as e:
    print(f"Error loading dataset: {e}")
```

## Outlier Visualization

We will generate boxplots for the following numerical columns:
- `satisfaction_level`
- `last_evaluation`
- `number_project`
- `average_montly_hours`
- `time_spend_company`

```python
df = df.select_dtypes(include=['number'])
```

```python
# Boxplot for satisfaction_level
plt.figure(figsize=(10, 6))
sns.boxplot(df)
plt.title('Boxplot of Satisfaction Level')
plt.show()
```
